```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Customer Churn.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#replaing blank with 0

df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['customerID'].duplicated().sum()
#no duplicate
```

```python
#Creating a customise function
def conv(value):
  if value == 1:
    return "yes"
  else:
      return "no"
df['SeniorCitizen']= df['SeniorCitizen'].apply(conv)
```

```python
df
```

```python
#Visualisation of customers churn
plt.figure(figsize=(4,4))
ax = sns.countplot( x = 'Churn',data = df)
ax.bar_label(ax.containers[0])
plt.title("count of customers churn")
plt.show()
```

```python

gb = df.groupby("Churn").agg({'Churn':"count"})
gb
plt.figure(figsize=(4,4))
plt.pie(gb['Churn'],labels = gb.index,autopct= "%1.2f%%")
plt.title("perchantage of churned customer",fontsize = 12)
plt.show()
```

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x = 'gender',data = df,hue='Churn',palette='viridis')
ax.bar_label(ax.containers[0])
plt.title("Churn rate by gender",fontsize = 12)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x = 'SeniorCitizen',data = df,hue='Churn',palette='viridis')
plt.title("Churn rate by SeniorCitizen",fontsize = 12)
plt.show()
```

```python
# Preprocessing data to calculate percentages
data = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='count')
total = data.groupby('SeniorCitizen')['count'].transform('sum')
data['percentage'] = (data['count'] / total) * 100

# Creating the stacked bar chart
fig, ax = plt.subplots(figsize=(6, 4))
senior_categories = data['SeniorCitizen'].unique()
colors = sns.color_palette('viridis', len(data['Churn'].unique()))

for i, churn_category in enumerate(data['Churn'].unique()):
    subset = data[data['Churn'] == churn_category]
    ax.bar(subset['SeniorCitizen'], subset['percentage'], label=churn_category,
           color=colors[i], bottom=data[data['Churn'] == data['Churn'].unique()[i - 1]]['percentage'].values if i > 0 else 0)

# Adding labels
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%")

# Customizing the plot
ax.set_title("Churn Rate by SeniorCitizen (Percentage)", fontsize=12)
ax.set_xlabel("SeniorCitizen")
ax.set_ylabel("Percentage")
ax.legend(title="Churn")
plt.show()
```

```python



columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the subplot grid
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))
axes = axes.flatten()  # Flatten the axes array for easier iteration

# Loop through columns and create a countplot for each
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i], palette='viridis',hue='Churn')
    axes[i].set_title(f'Countplot of {col}', fontsize=12)
    axes[i].set_xlabel('')  # Optional: remove x-axis label
    axes[i].set_ylabel('Count', fontsize=10)

# Remove any unused subplots if columns < axes
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```
